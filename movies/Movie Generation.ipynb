```python
master_storyteller = """You are a master movie creator. You write with
        Captivating narrative voice: Able to draw listeners in with a unique and engaging style of delivery.
Emotional intelligence: Skillfully evokes and manages the audience's emotions throughout the story.
Vivid imagery: Paints rich, detailed pictures with words, bringing scenes to life in the audience's mind.
Character depth: Creates complex, relatable characters with distinct personalities and motivations.
Pacing expertise: Knows when to build tension, when to reveal information, and how to maintain interest.
Adaptability: Can adjust the story on the fly based on audience reactions and engagement.
Thematic resonance: Weaves meaningful themes and universal truths into the narrative.
Mastery of structure: Crafts well-organized plots with satisfying arcs and resolutions.
Command of language: Uses rich vocabulary, figurative language, and varied sentence structures effectively.
Cultural awareness: Incorporates diverse perspectives and draws from a wide range of cultural references.
"""

def _text_following_bulleted_item(text: str) -> str:
    parts = text.split(". ")
    if len(parts) < 2:
        return ""
    return ". ".join(parts[1:])

def unique_id(size):
    chars = list(set(string.ascii_uppercase + string.digits).difference("LIO01"))
    return "".join(random.choices(chars, k=size))

def outline_list(outline):
    chapters = outline.split('\n')
    chapter_list = []
    for c in chapters:
        try:
            if len(c) > 0:
                chapter_list.append(c.split(': ')[1])
        except:
            pass
    return chapter_list


class MovieObject():
    def __init__(self, subject=''):
        self.subject = subject
        self.book_id = unique_id(7)
        self.system_prompt = ''
        self.characters = ''
        self.plot = ''
        self.setting = ''
        self.title = ''
        self.author = ''
        self.outline = {}
        self.outline_list = []
        self.overall_outline = ''
        self.chapters = []
        self.chapter_descriptions = []
        self.sub_chapters = []
        self.sub_chapter_descriptions = []
        self.book_text = []
        
    def save_to_json(self, filename):
        data = {
            'subject': self.subject,
            'system_prompt': self.system_prompt,
            'characters': self.characters,
            'plot': self.plot,
            'setting': self.setting,
            'title': self.title,
            'author': self.author,
            'outline': self.outline,
            'outline_list': self.outline_list,
            'overall_outline': self.overall_outline,
            'chapters': self.chapters,
            'chapter_descriptions': self.chapter_descriptions,
            'sub_chapters': self.sub_chapters,
            'sub_chapter_descriptions': self.sub_chapter_descriptions,
            'book_text': self.book_text
        }
        
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)

    @classmethod
    def load_from_json(cls, filename):
        with open(filename, 'r') as f:
            data = json.load(f)
        
        book = cls(subject=data['subject'])
        book.system_prompt = data['system_prompt']
        book.characters = data['characters']
        book.plot = data['plot']
        book.setting = data['setting']
        book.title = data['title']
        book.author = data['author']
        book.outline = data['outline']
        book.outline_list = data['outline_list']
        book.overall_outline = data['overall_outline']
        book.chapters = data['chapters']
        book.chapter_descriptions = data['chapter_descriptions']
        book.sub_chapters = data['sub_chapters']
        book.sub_chapter_descriptions = data['sub_chapter_descriptions']
        book.book_text = data['book_text']
        
        return book
        
    def character_creator(self):
        character_prompt = master_storyteller + f"""
        Create 10 intense characters for a movie about {self.subject}, detailing their relationships with one another.
        Ensure to create both men and women in a diverse world of characters.
        Heroes, antiheroes, and villians. Saints and martyrs, predators and victims, old and young.
        With every character, include their character development arc.
        At the center of each character is a simple, incepted truth, and a fatal flaw, each represented by a few words.
        Include their truth and their flaw.
        """
        self.characters = generate(character_prompt)
            
    def setting_creator(self):
        setting_prompt = master_storyteller + f"""
        Create 10 places in a setting for a movie about {self.subject}, which are where the action of the book takes place.
        """
        self.setting = generate(setting_prompt)
    
    def plot_creator(self):
        plot_prompt = master_storyteller + f"""
        Create 10 major plot points for a movie about {self.subject}, which are where the action of the movie takes place.
        The plot should be lively, indense, and filled with drama. Every plot point should be at least 5 sentences long,
        filled with multiple forms of conflict between the characters & a dramatic plot twist.
    
        The characters in the movie are: \n {self.characters}
        The settings in the movie are: \n {self.setting}
    
        Create a fantastic plot.
        """
        self.plot = generate(plot_prompt)

    def generate_suboutline(self, chapter, num_parts=None):
        num_parts = num_parts or random.randint(2, 3)
    
        
        outline_prompt = f"""
    
        The characters in the movie are: \n {self.characters}
        The settings in the movie are: \n {self.setting}
        The plot of the movie is: \n {self.plot}
            
        The overall movie outline is: \n{str(self.outline)}.
        
        Generate an {str(num_parts)} part subchapter outline about the scene{chapter}.
        Structure the outline to have the format:
        Subchapter x: <Title of Subchapter x>
        Subchapter y: <Title of Subchapter y>
        Subchapter z: <Title of Subchapter z>
    
        Without including any text outside of the clear subchapters text set out above.
        Do not repeat subchapter meanings titles that have already been used.
        """ 
        # print("Outline Prompt:", outline_prompt)
        outline = generate(outline_prompt)
        
        return outline_list(outline)[:num_parts]
    
    def generate_outline(self, num_parts=None):
        num_parts = num_parts or random.randint(10, 12)
        
        outline_prompt = f"""
    
        Generate an {str(num_parts)} part outline about {self.subject}.
    
    
        The characters in the movie are: \n {self.characters}
        The settings in the movie are: \n {self.setting}
        
        Structure the outline to have the format:
        Scene 1: <Title of Scene 1>
        Scene 2: <Title of Scene 2>
        Scene 3: <Title of Scene 3>
    
        Without including any text outside of the clear Scenes text set out above.
        """
        
        outline = generate(outline_prompt)
        self.outline_list = outline_list(outline)
        print("Generating Base Outline", flush=True)
        self.chapters = self.outline_list

    def generate_suboutlines(self):
        for c in self.chapters:
            print("Generating SubOutline", flush=True)
            sub_outline = self.generate_suboutline(c)
            self.sub_chapters.append(sub_outline)

    def scene_writer(self, chapter, subchapter):
        book_text = '\n'.join(self.book_text)
        chapter_prompt = f"""
    
        Previously written scene transcript:
        
        {book_text}
    
        The above is the text of the movie up until this point.
    
    
        Characters:
        {self.characters}
    
        Setting:
        {self.setting}
    
        Overall Book Outline:
        {self.overall_outline}
        
        Write a detailed, conflict rich, emotionally extreme movie scene, including gripping dialogue, in a style that is emotionally captivating. Be creative.
        Make sure the passage addresses content that is unique to its part of the outline. Do not focus on content addressed elsewhere in the outline.
        Do not include any headings or subheadings. Gracefully follow on from the previous text of this book and make all writing totally coherent with it.
        
        Write an *extremely long* phenomenal movie scene on {subchapter} inside the chapter called {chapter}.
        At least half (50%) of the text that you write *must be dialogue*.
    
        """
        subchapter_text = generate(chapter_prompt)
        print(subchapter_text)
        
        return subchapter_text
        

    def generate_text(self):
        self.overall_outline = ""
        for i, c in enumerate(self.chapters):
            self.overall_outline += f"Chapter {i+1}: {c}\n"
            try:
                for j, s in enumerate(self.sub_chapters[i]):
                    self.overall_outline += f"\tSubchapter {j}: {s}\n"
            except:
                pass
        
        for i, c in enumerate(self.chapters):
            if i < 1:
                subchapter_text = self.chapter_writer(c, "the chapter opener")
                self.book_text.append(subchapter_text)
                self.save_to_json(self.book_id + '_' + self.subject[:40])
                for j, s in enumerate(self.sub_chapters[i]):
                    subchapter_text = self.chapter_writer(c, s)
                    self.book_text.append(subchapter_text)
                    self.save_to_json(self.book_id + '_' + self.subject[:40])            
                    with open('/Users/jeremynixon/Dropbox/python_new/Misc/movies/%s'%(self.subject[:30] + '_' +  self.book_id), 'w') as f: f.write('\n'.join(self.book_text))

```

```python
def extract_characters(text):
    import re
    pattern = r'^\d+\.\s'
    lines = text.strip().split('\n')
    characters = []
    current_character = ""
    
    for line in lines:
        if re.match(pattern, line):
            if current_character:
                characters.append(current_character.strip())
            current_character = line
        else:
            current_character += "\n" + line
    
    if current_character:
        characters.append(current_character.strip())
    
    return characters

character_descriptions = extract_characters(mo.characters)[1:11]
```

```python
def generate_image(prompt, width, height, output_path):
    handler = fal_client.submit(
        "fal-ai/flux/dev",
        arguments={
            "prompt": prompt,
            "image_size": {
                "width": width,
                "height": height
            },
            "enable_safety_checker": False,
        },
    )
    
    result = handler.get()
    output_url = result['images'][0]['url']
    
    # Download and save the image
    response = requests.get(output_url)
    response.raise_for_status()
    
    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    
    # Save the image
    img = Image.open(BytesIO(response.content))
    img.save(output_path, "PNG")
    print(f"Generated: {output_path}")

def generate_character_images(character_descriptions):
    # Create base directory
    base_dir = "movies/assets"
    characters_dir = f"{base_dir}/characters"
    
    # Ensure characters directory exists
    os.makedirs(characters_dir, exist_ok=True)
    
    # List of visual descriptions for 10 characters
    # character_descriptions = [
    #     "elderly wizard with long white beard, glowing staff, blue robe, wise expression, pixel art style",
    #     "young female archer with red hair, green cloak, bow and arrow ready, determined look, pixel art style",
    #     "muscular barbarian warrior with animal fur armor, two-handed axe, wild hair, fierce expression, pixel art style",
    #     "mysterious rogue in dark hood, dual daggers, shadowy face, nimble pose, pixel art style",
    #     "armored knight in shining plate mail, shield with crest, longsword, noble stance, pixel art style",
    #     "graceful elven healer with flowing silver hair, white robes, staff with crystal, serene face, pixel art style",
    #     "dwarf engineer with leather apron, mechanical gadgets, thick beard, goggles on head, pixel art style",
    #     "tribal shaman with animal skull headdress, wooden staff, potion vials, colorful paint markings, pixel art style",
    #     "undead necromancer with skeletal hands, dark purple robes, spell book, glowing eyes, pixel art style",
    #     "monk with shaved head, simple cloth attire, prayer beads, meditative pose, calm expression, pixel art style"
    # ]
    
    # Standard character image dimensions
    width = 512
    height = 512
    
    # Generate all character images
    for i, description in enumerate(character_descriptions, 1):
        character_name = f"character_{i}"
        generate_image(
            prompt=description,
            width=width,
            height=height,
            output_path=f"{characters_dir}/{character_name}.png"
        )
        print(f"Generated character {i}: {description[:30]}...")
    
    print(f"All 10 character images have been generated in {characters_dir}")
```
